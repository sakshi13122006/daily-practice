# Bayes Theorem

Bayes Theorem is a mathematical formula used to determine conditional probability.

It helps update the probability of an event when new information or evidence becomes available.

It is widely used in statistics, artificial intelligence, machine learning, and medical diagnosis.

## Bayes Theorem Formula

The mathematical formula of Bayes Theorem is:

P(A|B) = (P(B|A) × P(A)) / P(B)

Where:

P(A) → Prior Probability  
P(B|A) → Likelihood  
P(B) → Evidence  
P(A|B) → Posterior Probability

## Terms in Bayes Theorem

1. Prior Probability (P(A))
   The initial probability of event A before new evidence is considered.

2. Likelihood (P(B|A))
   The probability of observing event B given that A has occurred.

3. Evidence (P(B))
   The total probability of observing event B.

4. Posterior Probability (P(A|B))
   The updated probability of A after considering evidence B.

## Example of Bayes Theorem

Suppose:

1% of people in a population have a disease.

A test has:
P(Positive | Disease) = 0.99

Probability that the test is positive in general:

P(Positive) = 0.05

We want to calculate:

Probability that a person actually has the disease given that the test is positive.

```python
# Bayes theorem calculation

P_disease = 0.01
P_positive_given_disease = 0.99
P_positive = 0.05

posterior_probability = (P_positive_given_disease * P_disease) / P_positive

print("Probability of disease given positive test:", posterior_probability)
```
