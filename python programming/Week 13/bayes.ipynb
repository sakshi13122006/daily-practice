## Introduction to Bayes Theorem

Bayes theorem is used to update probability when new information is available.

It helps calculate conditional probability.

## Bayes Theorem Formula

P(A|B) = (P(B|A) × P(A)) / P(B)

Where:

P(A) = Prior probability  
P(B|A) = Likelihood  
P(B) = Evidence  
P(A|B) = Posterior probability

```python
P_A = 0.01
P_B_given_A = 0.9
P_B = 0.05

bayes_probability = (P_B_given_A * P_A) / P_B

print("Posterior Probability: , ", bayes_probability)
```
