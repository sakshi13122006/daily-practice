## History of Probability

Probability theory originated in the 17th century from the study of gambling games.

Key contributors:

**Blaise Pascal (1623–1662)**  
Worked on probability problems related to gambling.

**Pierre de Fermat (1607–1665)**  
Collaborated with Pascal and helped develop early probability theory.

**Jacob Bernoulli (1654–1705)**  
Introduced the **Law of Large Numbers**.

**Pierre-Simon Laplace (1749–1827)**  
Expanded probability theory and applied it to scientific problems.

## What is Probability?

Probability measures the likelihood that an event will occur.

Range of probability:
0 ≤ P(E) ≤ 1

0 → Impossible event  
1 → Certain event

## Random Experiment

A random experiment is an experiment whose outcome cannot be predicted with certainty.

Examples:
- Tossing a coin
- Rolling a dice
- Drawing a card

## Sample Space

Sample space is the set of all possible outcomes of a random experiment.

Example:

Rolling a dice

S = {1,2,3,4,5,6}

```python
sample_space = [1,2,3,4,5,6]
print("Sample Space:", sample_space)
```

## Event

An event is a subset of the sample space.

Example:

Event A = getting an even number

A = {2,4,6}

```python
event_A = [2,4,6]
print("Event A:", event_A)
```

## Types of Events

1. Simple Event
2. Compound Event
3. Impossible Event
4. Certain Event
5. Complementary Event

## Complementary Event

If A is an event, then A' is its complement.

P(A) + P(A') = 1

## Classical Probability

Classical probability formula:

P(E) = Number of favourable outcomes / Total number of outcomes

```python
favourable = 3
total = 6

probability = favourable / total
print("Probability: ", probability)
```

## Empirical Probability

Empirical probability is calculated using experimental results.

P(E) = Number of times event occurs / Total trials

```python
import random

trials = 1000
heads = 0

for i in range(trials):
    if random.choice(["H","T"]) == "H":
        heads += 1

print("Experimental Probability:", heads/trials)
```

## Mutually Exclusive Events

Two events are mutually exclusive if they cannot occur at the same time.

Example:
Getting 2 and 5 in a single dice roll.

## Addition Rule of Probability

P(A ∪ B) = P(A) + P(B) − P(A ∩ B)

```python
P_A = 0.5
P_B = 0.4
P_intersection = 0.2

P_union = P_A + P_B - P_intersection
print("P(A union B):", P_union)
```

## Independent Events

Two events are independent if occurrence of one does not affect the other.

Example:
Coin toss and dice roll.

## Dependent Events

Two events are dependent if occurrence of one affects the probability of the other.

Example:
Drawing cards without replacement.

## Multiplication Rule

For independent events:

P(A ∩ B) = P(A) × P(B)

```python
P_A = 0.5
P_B = 0.2

P_AB = P_A * P_B
print("P(A and B):", P_AB)
```

## Union of Events

The union of two events A and B represents the event that either A occurs or B occurs or both occur.

It is written as:

A ∪ B

Example:
Getting an even number or a number greater than 4 when rolling a dice.

## Intersection of Events

The intersection of two events A and B represents the event that both A and B occur together.

It is written as:

A ∩ B

Example:
Probability that a card drawn is both red and a king.

```python
A = {2,4,6}
B = {4,5,6}

union = A.union(B)
intersection = A.intersection(B)

print("Union:", union)
print("Intersection:", intersection)
```

## Tree Diagram Method

A tree diagram is a graphical way to represent all possible outcomes of sequential events.

It is useful for solving probability problems involving multiple stages.

Example:
Two coin tosses.

```python
import itertools

outcomes = list(itertools.product(["H","T"], repeat=2))
print("Possible Outcomes:", outcomes)
```

## Conditional Probability

Conditional probability means probability of event A given that event B has occurred.

Formula:

P(A|B) = P(A ∩ B) / P(B)

```python
P_intersection = 0.2
P_B = 0.4

P_A_given_B = P_intersection / P_B
print("P(A|B):", P_A_given_B)
```
