EXEPTION HANDLING
 - An Exceptin is an runtime-error that occurs while execution of program. 
 - When an exception occurs, the normal flow of the program is interrupted.
 

Why we use it:
   -Without exception: chances of program crashes
   -With exeption: progran runs smoothly

Examples of exceptions:
    - Dividing a number by zero
    - Entering a string instead of a number
    - Accessing an invalid index of a list
    - Opening a file that does not exist


Exception handling is a mechanism used to detect, handle, and manage runtime errors so that the program does not crash and continues execution gracefully.

Python provides built-in support for exception handling using:

   1.try:
      - The try block contains the code that may cause an exception.

   2.except:
       - The except block handles the exception if it occurs.

   3.else:
       - The else block executes when no exception occurs.

   4.finally:
       - The finally block always executes, regardless of whether an exception occurs or not.


Syntax of Exception Hnadling:

try:
    # code that may cause exception
except ExceptionType:
    # handling code
else:
    # executes if no exception
finally:
    # executes always


Types of Exception Handling:

1. Built-in Exception Handling:
    - These are predefined exceptions in Python.
Examples:
     - ZeroDivisionError
     - ValueError
     - TypeError
     - IndexError
     - KeyError
     - FileNotFoundError

2. user-defined Exception Handling:
    - User-defined exceptions are created by programmers to handle specific error conditions in a program.

EXAMPLES OF BUILT-IN EXCEPTION HANDLING

```python
# Division by zero

try:
    a = int(input("Enter a number: "))
    b = int(input('Enter another number'))
    print("Result: ",a / b)
except ZeroDivisionError:
    print("Error: Cannot divide by Zero")
```

```python
#  ValueError

try:
    num = int(input("Enter a number: "))
    print('You entered: ',num)
except ValueError:
    print("Error: Please enter numbeers only")
```

```python
# Index error

try:
    lst = [10,20,30]
    print(lst[5])
except IndexError:
    print("Error: Index out of range")
```

```python
# Multiple Exceptions

try:
    x = int(input("enter x: "))
    y = int(input("Enter y: "))
    print(x / y)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Inavlid input : Enter Interger only ")
```

```python
# finally block

try:
    num = int(input("Enter a number: "))
    print(100/num)
except Exception as e:
    print("Error: ",e)
finally:
    print("Exception completed")
```

```python
# else block
try:
    n = int(input("Enter a number: "))
    print('Square: ',n*n)
except ValueError:
    print("Inavlid input")
else:
    print("program executed successfully")
    
```

```python
# generic Exception: Handle any error

try:
    a = int(input("Enter a number: "))
    b = int(input('Enter another number: '))
    print(a / b)
except Exception as e:
    print("error occurred: ",e) 
```

FILE HANDLING
- File handling allows a program to store data permanently in files and retrieve it later.

Types of files:
 1. text files: .txt, .csv, .log
 2. binary files: images, audio, video

steps in file Handling
 1. open a file
 2. perform operation(read/ write /   append)
 3. close the file


Opening a File:
       file = open("filename","mode")

File Modes:
1. r -> Read file
2. w -> Write file(overwrite)
3. a -> Append file
4. r+ -> Read + Write
5. w+ -> Write + Read
6. a+ -> Append + Read


Closing a File:
    file.close()


```python
# Write to a File(w)

file = open("data.txt","w")
file.write("Hello Sakshi\n")
file.write("File handling")
file.close()
```

```python
# Read from a file(r)

file = open("data.txt")
print(file.read())
file.close()
```

```python
# Append to a File(a)

file = open("data.txt","a")
file.write("\nThis line is Append")
file.close()
```

```python
file = open("data.txt")
print(file.read())
file.close()
```

```python
# File handling with exception handling

try:
    file = open("data.txt","r")
    print(file.read())
except FileNotFoundError:
    print("File not found")
finally:
    file.close()
```

```python
# store student data 

with open("student.txt","a") as file:
    name = input("Enter name: ")
    marks = input("Enter marks: ")
    file.write(name + " "+ marks + "\n")
```

```python
# read data

file = open("student.txt")
print(file.read())
file.close
```

```python
# read data

with open("student.txt","r") as file:
    for line in file:
        print(line.strip())
```
