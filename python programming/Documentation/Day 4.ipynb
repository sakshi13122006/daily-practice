FUNCTION

A function in Python is a reusable block of code that performs a specific task.
Functions help in code reuse, readability, and easy maintenance.

Types of functions:
 - Built-in functions
 - user defined functions
 - lambda function

 Syntax:
 
 def function_name(parameters):
    """Docstring: explains what the function does"""
    statements
    return value


```python
def add(a, b):
    """Returns sum of two numbers"""
    return a + b

result = add(10, 20)
print(result)
```

```python
'''Built in function example'''

print("Hello Sakshi")
```

```python
'''
A lambda function is a small anonymous function used for short operations.

Syntax:
lambda arguments : expression

'''

square = lambda x: x * x
print(square(5))
```

USER-DEFINED FUNCTIONS

A user defined function is a function created by the user to perform a specific task.
It helps in code reusability, readability, and modular programming.



```python
# programs of user-defined functions

def add(a,b):
    return a + b
    return result

print(add(10,20))
```

```python
# Function without parameter
def greet():
    print("hello Sakshi")

greet()
```

```python
# Function with parameters

def square(n):
    return n*n

print(square(5))
```

```python
# Function with return value

def area_of_circle(r):
    return 3.14 * r * r

print(area_of_circle(7))
```

```python
# Function without return value

def message():
    print("Python")

message()
```

```python
# Function call flow

def multiply(a , b):
    return a * b

result = multiply(4, 5)
print(result)
```

```python
def check_year(year):
    if year % 4 == 0:
        return "Leap year"
    else:
        return "Not a leap year"
    
y = int(input("Enter year to check: "))
print(y,"is",check_year(y))
```

```python
# Lambda
add = lambda a, b: a + b
print(add(5,3))
```

```python
# even odd example using lambda

even_odd = lambda n: "even" if n % 2 == 0 else "odd"
print(even_odd(10))
print(even_odd(7))
```

```python
square = lambda x: x**2
print(square(4))
```
