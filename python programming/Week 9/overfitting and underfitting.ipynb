1. Underfitting

- Definition: The model is too simple to capture the underlying patterns in the data.

- Problem: It performs poorly on both training and test data.

Signs:

- Low accuracy on training data.
- Cannot capture trends, just general guesses.

Causes:

 - Model is too simple (e.g., linear model for complex data).
 - Not enough features or data.

Example: 
  - Trying to fit a straight line to a curve.





2. Overfitting

- Definition: The model is too complex, memorizes the training data, including noise.

- Problem: Performs very well on training data but poorly on new/test data.

Signs:

 - High accuracy on training data.
 - Low accuracy on unseen/test data.

* Causes:

  - Too complex model (e.g., deep neural network on small dataset).
  - Too many features.
  - No regularization.

Example: A wiggly curve that touches all training points perfectly but fails on new data.

* Visualization Idea

Imagine a scatter plot of points:
  - Underfit: The line is flat, missing trends.
  - Overfit: The line twists through every point.
  - Just Right (Good fit): The line follows the overall trend without touching every point.
