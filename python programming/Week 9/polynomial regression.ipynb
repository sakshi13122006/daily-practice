Polynomial Regression Documentation
1. Introduction

Polynomial Regression is an extension of Linear Regression. Instead of fitting a straight line, it fits a polynomial curve to the data. It is useful when the relationship between independent variable 
𝑋 and dependent variable 
𝑌 is non-linear.

Equation




2. When to Use Polynomial Regression

 - The data shows a curved relationship.
 - Linear regression underfits the data.

Examples:

 - Predicting growth rates,
 - Modeling temperature trends,
 - Forecasting population growth.



3. Advantages;

 - Can model non-linear relationships.
 - Simple extension of linear regression.
 - Easy to implement.

4. Disadvantages

 - Can overfit if degree is too high.
 - Sensitive to outliers.
 - Choosing the right degree is critical.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("salary.csv")
df.shape
```

```python
df.columns
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x, y)
lr.score(x, y)
```

```python
# polynomial
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 13)
```

```python
x_poly = poly.fit_transform(x)
```

```python
x_poly
```

```python
lr.fit(x_poly, y)
```

```python
lr.score(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lr.predict(x_poly))
plt.grid()
```
