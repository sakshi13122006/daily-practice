What is Regularization?

- Regularization is a technique in machine learning used to prevent overfitting by penalizing large or complex model parameters. Basically, it helps the model stay simple and generalize better to new data.

Why Do We Need It?

- Overfitting happens when a model memorizes training data.
- Regularization adds a penalty term to the loss function to discourage overly complex models.

Types of Regularization

1️⃣ L1 Regularization (Lasso)

 - Adds sum of absolute values of weights to the loss function.

Effects:

 - Can shrink some weights to exactly 0 → feature selection.
 - Makes model simpler.


 2️⃣ L2 Regularization (Ridge)

  - Adds sum of squared weights to the loss function.

Effects:

   - Shrinks weights but rarely to 0.
   - Smooths the model, prevents huge coefficients.

3️⃣ Elastic Netrmula:

  - Combination of L1 + L2.
  - Good when you want feature selection + smoothness.
