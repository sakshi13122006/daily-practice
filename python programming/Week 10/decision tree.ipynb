1️⃣ Definition

- A Decision Tree (DT) is a supervised learning algorithm used for classification and regression tasks.
- It splits data into branches based on feature values to make predictions.
- The structure looks like a tree:
- Root Node: First decision point
- Internal Nodes: Feature tests
- Leaf Nodes: Final class label or value

Example: Predict whether a person buys a computer based on age, income, and student status.


2️⃣ Key Concepts

  - Node: Point where data is split (Root, Internal, Leaf).
  - Edge / Branch: Outcome of a test, connecting nodes.
  - Splitting: Dividing a node into sub-nodes based on a feature.
  - Impurity / Homogeneity Metrics: How “mixed” the node is. Common ones:
          - Gini Index
          - Entropy / Information Gain
          - Variance (for regression)

  - Pruning: Removing branches that don’t improve performance to reduce overfitting.
  - Depth: Number of levels in the tree; deeper trees → complex, may overfit.
  

3️⃣ Decision Tree Algorithm Workflow

  1. Select Best Feature: Based on impurity reduction (Gini or Entropy). 
  2. Split Dataset: Divide data according to feature’s possible values.
  3. Repeat for Sub-nodes: Continue splitting until:
       - All nodes are pure (single class)
       - Max depth is reached
       - Minimum samples per node are reached

  4. Assign Leaf Values: Final prediction based on majority class (classification) or mean value (regression).


4. Advantages:

   - Easy to understand and interpret (like a flowchart).
   - Can handle numerical and categorical data.
   - No scaling required (unlike SVM or KNN).


5. Disadvantages:

   - Prone to overfitting, especially deep trees.
   - Sensitive to small changes in data.
   - Can be biased towards features with more levels.

```python
# Example Code (Python – Classification)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Sample dataset
X = [[25,50000], [35,60000], [45,80000], [20,30000], [40,90000]]
Y = [0,1,1,0,1]  # 0 = No, 1 = Yes

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Feature scaling (optional for DT)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Train Decision Tree
dt_classifier = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
dt_classifier.fit(x_train_scaled, y_train)

# Predict
y_pred = dt_classifier.predict(x_test_scaled)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Visualization of Decision Tree

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plot_tree(dt_classifier, feature_names=['Age','Income'], class_names=['No','Yes'], filled=True)
plt.show()
```
