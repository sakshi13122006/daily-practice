Classification in Machine Learning – Complete Documentation


1. Definition

 - Classification is a supervised learning task where the goal is to predict the category or class label of given data points based on input features.

  - Input: Features (numerical or categorical)

  - Output: Class label (categorical)

Example: Predict whether an email is Spam or Not Spam.


2. Key Concepts:

  - Features (X): Input data used for prediction.
  - Labels (Y): The target categories/classes.
  - Training set: Data used to train the model.
  - Testing set: Data used to evaluate model performance.
  - Classifier: The algorithm or model that performs classification (e.g., Logistic Regression, Decision Tree).



3. Types of Classification:

a) Binary Classification

  - Only two classes.
  - Example: Email → Spam / Not Spam

b) Multi-class Classification

   - More than two classes.
   - Example: Handwritten digit recognition → 0,1,2,…,9

c) Multi-label Classification

   - Each sample can belong to multiple classes at the same time.
   - Example: News article → Politics + Technology


4. Classification Workflow:

  - Data Collection: Gather dataset with features and labels.
  - Data Preprocessing: Handle missing values, encode categorical variables, normalize/scale features.
  - Train-Test Split: Divide data into training and testing sets.
  - Choose Classifier: Select algorithm (e.g., Logistic Regression, SVM, Random Forest).
  - Train Model: Fit the classifier on training data.
  - Make Predictions: Predict class labels on testing data.
  - Evaluate Model: Use metrics like Accuracy, Precision, Recall, F1-Score, ROC-AUC.
  - Deploy Model: Use in real applications once satisfied with performance.

6. Evaluation Metrics

  - Accuracy: Correct predictions / Total predictions
  - Precision: Correct positive predictions / Total predicted positives
  - Recall (Sensitivity): Correct positive predictions / Total actual positives
  - F1-Score: Harmonic mean of Precision and Recall
  - Confusion Matrix: Shows True Positives, True Negatives, False Positives, False Negatives
  - ROC Curve & AUC: Measures classifier’s ability to distinguish classes

```python
# Example Code (Python – Logistic Regression)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Sample dataset (X = features, Y = labels)
X = [[1,2], [2,3], [3,4], [4,5]]
Y = [0, 0, 1, 1]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

# Feature scaling
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Train classifier
classifier = LogisticRegression()
classifier.fit(x_train_scaled, y_train)

# Predict
y_pred = classifier.predict(x_test_scaled)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

8. Applications of Classification
