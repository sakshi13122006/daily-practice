Module in Python:

A module is a file containing Python code (functions, classes, variables) that you can reuse in other Python programs.

Module = reusable code

Helps organize programs into smaller pieces

Can be built-in, external, or user-defined


```python
# maath - math function
import math

print(math.sqrt(25))
print(math.factorial(5))
```

```python
# datetime - date and time

import datetime

now = datetime.datetime.now()
print(now)
```

```python
# user defined module

import math_ex

print(math_ex.add(10,20))
print(math_ex.sub(40,10))
print(math_ex.multiply(5,5))
```
