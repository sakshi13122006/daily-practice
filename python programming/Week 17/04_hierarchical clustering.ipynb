# What is Hierarchical Clustering?

Hierarchical Clustering is an unsupervised learning technique used to group similar data points.

It builds clusters step-by-step in a tree-like structure.

## Key Idea
Clusters are formed by merging or splitting groups of data points.

# Types of Hierarchical Clustering

## 1. Agglomerative (Bottom-Up)
- Start with each point as a separate cluster
- Merge closest clusters step by step

## 2. Divisive (Top-Down)
- Start with one big cluster
- Split into smaller clusters

👉 Most commonly used: Agglomerative Clustering

# How Agglomerative Clustering Works

1. Treat each data point as a cluster
2. Find the two closest clusters
3. Merge them
4. Repeat until only one cluster remains

This process creates a hierarchy (tree structure).

# What is a Dendrogram?

A dendrogram is a tree diagram used to visualize hierarchical clustering.

## It shows:
- how clusters are merged
- at what distance they merge

## Important
Cutting the dendrogram at a certain height gives the number of clusters.

```python
# Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

# Step 1: Create Dataset

We create sample data using make_blobs for clustering.

```python
# Create dataset

X, _ = make_blobs(n_samples=200, centers=4, random_state=42)

print("Shape:", X.shape)
```

```python
# Plot original data

plt.scatter(X[:,0], X[:,1])
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# Program Understanding: Dataset

- 200 data points are created
- Data is grouped around 4 centers
- This helps in understanding clustering clearly

# Step 2: Create Dendrogram

We use linkage() to compute cluster merging and dendrogram() to plot it.

```python
# Create linkage matrix

linked = linkage(X, method='ward')

# Plot dendrogram

plt.figure(figsize=(10,5))
dendrogram(linked)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

# Program Explanation: Dendrogram

- linkage() calculates how clusters should be merged
- method='ward' minimizes variance within clusters
- dendrogram() draws tree structure

## Interpretation
- Lower merges → points are closer
- Higher merges → clusters are far apart

Cutting the tree at a height gives number of clusters.

# Step 3: Apply Agglomerative Clustering

Now we apply clustering using sklearn.

```python
# Apply Agglomerative Clustering

model = AgglomerativeClustering(n_clusters=4)
labels = model.fit_predict(X)

print("Cluster labels:", labels[:10])
```

```python
# Plot clusters

plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Hierarchical Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

# Program Explanation: Clustering

- AgglomerativeClustering creates clusters
- n_clusters=4 means we want 4 groups
- fit_predict() assigns cluster labels

## Result
Points with same color belong to same cluster.

# Linkage Methods

Different ways to measure distance between clusters:

## 1. Single Linkage
Minimum distance between points

## 2. Complete Linkage
Maximum distance between points

## 3. Average Linkage
Average distance

## 4. Ward Linkage
Minimizes variance (most commonly used)

# Advantages of Hierarchical Clustering

1. No need to choose number of clusters initially
2. Easy to understand using dendrogram
3. Works well for small datasets
4. Provides hierarchy of clusters

# Disadvantages

1. Slow for large datasets
2. Cannot undo merging
3. Sensitive to noise
4. Memory intensive

# Applications

- Customer segmentation
- Document clustering
- Image segmentation
- Gene analysis
- Market research

# Comparison with K-Means

| Feature | Hierarchical | K-Means |
|--------|-------------|--------|
| Structure | Tree | Flat clusters |
| Need K | Not initially | Required |
| Speed | Slow | Fast |
| Visualization | Dendrogram | Scatter plot |

# Conclusion

Hierarchical clustering is a powerful clustering technique that builds a tree structure of data.

It is easy to understand and visualize but slower than K-Means.

Hence, it is best suited for small datasets and analysis tasks.
