# Introduction to Association Rule Mining

Association Rule Mining is a data mining technique used to find relationships between items in a dataset.

It helps answer questions like:
- Which items are purchased together?
- If a customer buys one item, what other item is likely to be bought?

## Example
If customers often buy:
- Bread and Butter together

Then we can write the rule:
Bread → Butter

This means customers who buy Bread are likely to buy Butter.

## Applications
- Market basket analysis
- Product recommendation
- Retail analysis
- Web usage mining

# Important Terms

## 1. Item
A single product or object

Example:
Milk

## 2. Itemset
A group of items

Example:
{Milk, Bread}

## 3. Rule
A relationship written in the form:

A → B

This means:
If A is present, B is likely to be present

## 4. Frequent Itemset
An itemset that appears frequently in the dataset

# Measures Used in Association Rule Mining

There are three important measures:

## 1. Support
Support tells how frequently an itemset appears in the dataset.

Support(A) = (Number of transactions containing A) / (Total number of transactions)

## 2. Confidence
Confidence tells how often B appears when A is already present.

Confidence(A → B) = Support(A ∪ B) / Support(A)

## 3. Lift
Lift tells the strength of the rule.

Lift(A → B) = Confidence(A → B) / Support(B)

## Interpretation of Lift
- Lift > 1 → positive relation
- Lift = 1 → no relation
- Lift < 1 → negative relation

# What is Apriori Algorithm?

Apriori is a popular algorithm used to find frequent itemsets in transaction data.

## Main Idea
If an itemset is frequent, then all of its subsets must also be frequent.

## Steps of Apriori
1. Find frequent single items
2. Generate larger itemsets from smaller frequent itemsets
3. Remove itemsets with low support
4. Repeat until no more frequent itemsets are found

```python
# Import required libraries

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

# Step 1: Create Transaction Dataset

We create a simple market basket dataset.

Each row represents one transaction.
Each transaction contains the items purchased together.

```python
# Sample transactions

transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Butter'],
    ['Milk', 'Bread']
]

print("Total transactions:", len(transactions))
transactions
```

# Program Understanding: Transaction Data

In this dataset:
- Each list is one transaction
- Items inside the list are bought together

## Example
['Milk', 'Bread', 'Butter']

This means the customer bought:
- Milk
- Bread
- Butter
in the same transaction

# Step 2: Convert Transactions into Binary Table

Machine learning algorithms cannot directly work on list format.

So we convert the transaction data into a binary table:
- True means item is present
- False means item is absent

```python
# Convert transactions into binary format

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)
df
```

# Program Explanation: Encoded Table

In this table:
- Each row is one transaction
- Each column is one item
- True means the item is present in that transaction
- False means the item is not present

This format is required for the Apriori algorithm.

# Step 3: Find Frequent Itemsets using Apriori

Now we apply the Apriori algorithm.

## Parameter Used
- `min_support=0.3`

This means an itemset must appear in at least 30% of transactions to be considered frequent.

```python
# Apply Apriori algorithm

frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)

frequent_itemsets
```

# Program Explanation: Frequent Itemsets

The output shows:
- `support` → frequency of the itemset
- `itemsets` → set of items found frequently

## Example
If itemset {Bread, Butter} has support 0.57,
it means 57% of transactions contain Bread and Butter together.

# Step 4: Generate Association Rules

Now we generate rules from the frequent itemsets.

## Parameter Used
- `metric="confidence"`
- `min_threshold=0.6`

This means we only keep rules whose confidence is at least 60%.

```python
# Generate association rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

rules
```

```python
# Display only important columns

rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

# Program Explanation: Association Rules

In the rules table:

- `antecedents` → left side of rule (A)
- `consequents` → right side of rule (B)
- `support` → how frequently rule appears
- `confidence` → how likely B appears when A appears
- `lift` → strength of the rule

## Example
If rule is:

{Milk} → {Bread}

It means customers who buy Milk are also likely to buy Bread.

# Example Interpretation

Suppose one rule is:

{Milk} → {Bread}

and:
- confidence = 0.75
- lift = 1.05

## Meaning
- 75% of customers who buy Milk also buy Bread
- Since lift is greater than 1, the relationship is positive

# Why Support, Confidence, and Lift are Important

## Support
Helps us know whether the rule is common enough

## Confidence
Helps us know how reliable the rule is

## Lift
Helps us know whether the relationship is actually strong or just accidental

# Difference Between Frequent Itemset and Rule

## Frequent Itemset
A group of items that appears frequently

Example:
{Milk, Bread}

## Association Rule
A relation between itemsets

Example:
Milk → Bread

## Simple Difference
- Frequent itemset = common group
- Rule = directional relationship

# Memory Trick

- Support = How often
- Confidence = How sure
- Lift = How strong

## Final Memory Line
Support checks frequency, confidence checks probability, and lift checks strength.
