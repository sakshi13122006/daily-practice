K-Nearest Neighbour (KNN)
📌 What is KNN?
    - K-Nearest Neighbour (KNN) is a supervised machine learning algorithm used for:
                - Classification
                - Regression
    - It predicts the output by looking at the K closest data points (neighbors).

⚙️ How it Works

    1. Choose value of K
    2. Calculate distance from test point to all training points
    3. Select K nearest neighbors
    4. Classification → majority vote
       Regression → average value

 

📊 Types:
🔹 Classification:
    - Output → Category
    - Example: Spam/Not Spam

🔹 Regression:
    - Output → Numeric value
    - Example: Price prediction

🎯 Choosing K:
    - Small K → Overfitting
    - Large K → Underfitting
	​

✅ Advantages:
     - Simple
     - No training time
     - Easy to understand

❌ Disadvantages:
     - Slow for large data
     - Needs feature scaling
     - Memory heavy



```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('ingredients.csv')
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.jointplot(x = 'sweetness', y ='crunchiness', data = df,
             hue = 'class')
```

```python
x = df[['sweetness', 'crunchiness']]
y = df['class']
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 4)
knn.fit(x, y)
```
