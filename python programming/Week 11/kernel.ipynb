# Support Vector Machine (SVM) – Kernel Functions

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression.

SVM separates classes using an optimal hyperplane.  
When data is not linearly separable, kernel functions are used to transform data into higher dimensions.

This notebook demonstrates:
- Linear Kernel
- Polynomial Kernel
- RBF Kernel
- Sigmoid Kernel

## What is a Kernel?

A kernel is a mathematical function that transforms input data into a higher dimensional space
so that it becomes easier to separate classes.

It helps SVM handle non-linear data.

Kernel Trick:
Instead of computing transformation explicitly, SVM uses kernel functions to calculate it efficiently.

## Types of SVM Kernels

## 🔹 Linear Kernel

### Definition
Linear kernel is the simplest kernel used in SVM.  
It draws a straight line (or hyperplane) to separate classes.

### Formula
K(x, y) = x · y

### Working
It calculates the dot product between two feature vectors.
If data is already linearly separable, no transformation is needed.

### Advantages
- Very fast training
- Less memory usage
- Works well for high dimensional data
- Simple to understand

### Disadvantages
- Cannot handle complex or curved boundaries
- Poor performance for non-linear data

### When to Use
- Linearly separable data
- Text classification
- Small or medium datasets

### Example Applications
- Spam detection
- Sentiment analysis
- Basic binary classification


## 🔹 Polynomial Kernel

### Definition
Polynomial kernel creates curved decision boundaries by transforming features into higher dimensions.

### Formula
K(x, y) = (x · y + 1)^d

where d = degree of polynomial

### Working
It considers combinations of features and their interactions.
Higher degree creates more complex curves.

### Advantages
- Handles moderately complex data
- More flexible than linear kernel
- Captures feature relationships

### Disadvantages
- Slower than linear
- Can overfit for large degree values
- More parameters to tune

### When to Use
- Data with curved boundaries
- Medium complexity problems

### Example Applications
- Pattern recognition
- Image classification

## 🔹 RBF (Radial Basis Function) Kernel

### Definition
RBF kernel is the most popular and powerful kernel.
It maps data into infinite dimensional space to handle complex patterns.

### Formula
K(x, y) = exp(-γ ||x − y||²)

where γ (gamma) controls influence of each point

### Working
It measures similarity based on distance.
Closer points → higher similarity
Far points → lower similarity

### Advantages
- Very high accuracy
- Works for both linear and non-linear data
- Default choice for most problems
- Handles complex decision boundaries

### Disadvantages
- Slower training time
- Requires parameter tuning (C and γ)
- More computational cost

### When to Use
- Complex datasets
- Unknown data distribution
- Most real-world ML problems

### Example Applications
- Handwriting recognition
- Face recognition
- Medical diagnosis

## 🔹 Sigmoid Kernel

### Definition
Sigmoid kernel behaves similar to a neural network activation function.

### Formula
K(x, y) = tanh(α x · y + c)

### Working
It transforms data using hyperbolic tangent function,
similar to neurons in Artificial Neural Networks.

### Advantages
- Works like neural network
- Can handle some non-linear data

### Disadvantages
- Less stable
- Not widely used
- Lower accuracy compared to RBF
- Harder to tune

### When to Use
- Experimental purposes
- Neural-network-like behavior

### Example Applications
- Rarely used in practice
- Research experiments

## Kernel Comparison

| Kernel      | Speed | Accuracy | Use Case |
|------------|--------|-----------|-----------|
| Linear     | Fast   | Medium    | Linearly separable data |
| Polynomial | Medium | Good      | Curved boundaries |
| RBF        | Slow   | Very High | Complex patterns |
| Sigmoid    | Medium | Medium    | Neural-like behavior |

```python
# linear

import pandas as pd
from sklearn import svm
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv("linear_data.csv")

X = df[['x1','x2']]
y = LabelEncoder().fit_transform(df['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
# poly

df = pd.read_csv("poly_data.csv")

X = df[['x1','x2']]
y = LabelEncoder().fit_transform(df['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = svm.SVC(kernel='poly', degree=2)
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
# rbf

df = pd.read_csv("rbf_data.csv")

X = df[['x1','x2']]
y = LabelEncoder().fit_transform(df['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = svm.SVC(kernel='rbf')
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
# sigmoid

df = pd.read_csv("sigmoid_data.csv")

X = df[['x1','x2']]
y = LabelEncoder().fit_transform(df['label'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = svm.SVC(kernel='sigmoid')
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
for k in ['linear','poly','rbf','sigmoid']:
    model = svm.SVC(kernel=k)
    model.fit(X_train, y_train)
    acc = accuracy_score(y_test, model.predict(X_test))
    print(k, ":", acc)
```
