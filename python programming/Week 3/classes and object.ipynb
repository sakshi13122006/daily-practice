OOP Concepts In Python
OOP (Object Oriented Programming) is a programming style where we create objects that represent real-world things.

Python supports OOP to make code:
  - reusable 
  - organized 
  - secure     
  
 Concepts of OOP:
 
  1. class
      - A class is a collection of objects. Classes are blueprints for creating objects. A class defines a set of attributes and methods that the created objects (instances) can have.
      - Secure 
      
    Some points on Python class:  
      - Classes are created by keyword class.
      - Attributes are the variables that belong to a class.
      - Attributes are always public and can be accessed using the dot (.) operator. Example: Myclass.   Myattribute                                                               

```python
class Employee:   # class
    # name = "Ash"
    language = "python"  # This is a class attribute
    salary = 350000


ash = Employee()
ash.name = "Ash"  # This i as an Instance Attribute
print(ash.name, ash.language, ash.salary)

roy = Employee()
roy.name = "Roy"
print(roy.name, roy.salary, roy.language)

#  Here name is Instance attribute and language and salary are class attribute as they directly belong to the class
```

```python
# Instance attributes take preference over class attributes during assignment and retrival. 

class emp:
     lang = "Python"
     salary = 25000

la = emp()
la.lang = "Java"
print(la.lang, la.salary)
```

```python
# self
# class is written once but objects are created many times
# when we create object python must know "For WHICH object I store or read data" that's why self exits..

class emp:
    lang = "Python"
    salary = 25000

    def getInfo(self):
        print(f"The language is {self.lang} and the salary is {self.salary}")

    def greet(self):
        print("Good Morning")

ash = emp()
ash.greet()
ash.getInfo()
```

```python
# @staticmethod

class emp:
    lang = "Python"
    salary = 25000

    def getInfo(self):
        print(f"The language is {self.lang} and the salary is {self.salary}")

    @staticmethod
    def greet():
        print("Good Morning")

ash = emp()
ash.greet()
ash.getInfo()
```

```python
# @staticmethod
# It is used when a function belongs to a clss logically, but does not need object,class. In simple language It is related to the class, but independent of class data and object data.txt

class student:
    @staticmethod
    def greet():
        print("Hello, Good morning")

student.greet()
```

```python
# Example 1 (Create class to store info of few programmers)

class programmer:
    company = "Microsoft"
    def __init__(self, name,salary, pin):
        self.name = name
        self.salary = salary
        self.pin = pin

p = programmer("Ashu",120000,411000)
print(p.name,p.salary,p.pin,p.company)

r = programmer("ria",300000, 450043)
print(r.name, r.salary, r.pin)

```
