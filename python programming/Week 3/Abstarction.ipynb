1. ABSTRACTION:
   - Abstraction means showing only the essential features of an object and hiding the internal implementation details.

   - In simplr words: tell what an object can do, not how it does it.


2. Need Of Abstraction:
  - Reduce Complexity
  - Improve code readability
  - Enforce a common structure
  - Make code easy to maintain and extend
  - it is especially useful in large projects where many developers work together.


3. Real-Life Example:
  - When we use a mobile phone:
     - We know what it can do (call, camera, internet)
     - We donâ€™t know how it works internally
  - This concept is abstraction.


4. Abstarction In Python:
    - Python provides abstraction using:
         - Abstract Base Classes (ABC)
         - Abstract methods

    - To use abstraction, Python provides the abc module.


5. Abstarct Base Class(ABC):
   - An abstract class:
     - Acts as a blueprint
     - Cannot be instantiated (object cannot be created)
     - Contains abstract methods

  - Syntax:
    from abc import ABC

    class ClassName(ABC):
       pass


6. Abstract Method:
   - An Abstract Method:
      - Is declared but not implemented
      - Must be implemented by the child class.

  - Syntax:
  from abc import abstractmethod
 
  @abstractmethod
  def method_name(self):
       pass


7. Important Rules of Abstraction:

   -  Object of an abstract class cannot be created
   - Child class must implement all abstract methods
   - If not implemented, Python raises an error






```python
# Example of abstraction

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class dog(Animal):
    def sound(self):
        print("Dog barks")

class cat(Animal):
    def sound(self):
        print("Cat meows")

d = dog()
d.sound()

c = cat()
c.sound()
```

```python
from abc import ABC, abstractmethod

class phone(ABC):

    @abstractmethod
    def camera(self):
        pass

class samsung(phone):
    def camera(self):
        print("50MP Camera")

class iphone(phone):
    def camera(self):
        print("48MP camera")

s = samsung()
s.camera()

i = iphone()
i.camera()
```
