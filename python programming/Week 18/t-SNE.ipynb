# Introduction to t-SNE

t-SNE is a dimensionality reduction technique used mainly for visualization.

It converts high-dimensional data into low-dimensional data, usually:
- 2 dimensions
- 3 dimensions

## Full Form
t-SNE = t-Distributed Stochastic Neighbor Embedding

## Main Purpose
To visualize complex data in a simpler form.

## Important Point
t-SNE is mainly used for visualization, not for prediction.

# Why Do We Need t-SNE?

Real-world datasets may have many features.

Example:
- 10 features
- 50 features
- 100 features

Such data is difficult to visualize directly.

t-SNE helps us:
- reduce dimensions
- preserve local structure
- show similar data points close together

# Main Idea of t-SNE

t-SNE tries to place similar data points near each other in lower dimensions.

## Simple Meaning
- Similar points stay close
- Dissimilar points stay far apart

It is very useful when we want to see clusters or groups in data.

# Difference Between PCA and t-SNE

| Feature | PCA | t-SNE |
|---------|-----|------|
| Type | Linear | Non-linear |
| Goal | Reduce variance | Preserve neighborhood |
| Speed | Faster | Slower |
| Use | Compression + visualization | Mainly visualization |

## Simple Difference
- PCA keeps maximum variance
- t-SNE keeps similar points close

```python
# Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
```

# Step 1: Load Dataset

We use the digits dataset.

## About Dataset
It contains handwritten digit images:
- 0 to 9

Each image is stored as numerical features.

```python
# Load dataset

digits = load_digits()
X = digits.data
y = digits.target

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# Convert to DataFrame for better understanding

df = pd.DataFrame(X)
df.head()
```

# Program Understanding: Dataset

In this dataset:
- Each row is one digit image
- Each column is one feature
- Total features = 64
- Target values = digits 0 to 9

This is high-dimensional data, so it is difficult to visualize directly.

# Step 2: Apply t-SNE

Now we reduce 64 features to 2 dimensions.

## Why 2 dimensions?
Because 2D data can be easily plotted on a graph.

```python
# Apply t-SNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

print("Shape after t-SNE:", X_tsne.shape)
```

# Program Explanation: t-SNE

In this code:
- `TSNE(n_components=2)` creates a t-SNE model for 2D output
- `random_state=42` gives reproducible result
- `fit_transform(X)` reduces the dimensions

## Result
The original 64-feature data becomes 2-feature data.

```python
# Plot t-SNE result

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.title("t-SNE Visualization of Digits Dataset")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.colorbar(scatter)
plt.show()
```

# Result Understanding

In this graph:
- Each point represents one digit image
- Points with similar pattern are grouped together
- Different colors represent different digits

## Meaning
If points form groups, it means t-SNE has preserved similarity well.

# Important Parameters of t-SNE

Some common parameters are:

## 1. n_components
Number of output dimensions

Example:
- 2 for 2D
- 3 for 3D

## 2. perplexity
Controls how t-SNE balances local and global structure

Common values:
- 5 to 50

## 3. random_state
Used for reproducibility

## 4. learning_rate
Controls optimization speed

# Important Parameters of t-SNE

Some common parameters are:

## 1. n_components
Number of output dimensions

Example:
- 2 for 2D
- 3 for 3D

## 2. perplexity
Controls how t-SNE balances local and global structure

Common values:
- 5 to 50

## 3. random_state
Used for reproducibility

## 4. learning_rate
Controls optimization speed

```python
# Apply t-SNE with a custom perplexity

tsne2 = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne2 = tsne2.fit_transform(X)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_tsne2[:, 0], X_tsne2[:, 1], c=y)
plt.title("t-SNE with Perplexity = 30")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.colorbar(scatter)
plt.show()
```

# Practical Observation

Different parameter values may give different plots.

This is normal in t-SNE.

## Important Note
t-SNE is sensitive to parameters like:
- perplexity
- learning rate
- random seed

# Advantages of t-SNE

1. Very useful for visualization
2. Preserves local relationships
3. Good at showing clusters
4. Works well on high-dimensional data
5. Helps in exploratory data analysis

# Disadvantages of t-SNE

1. Slow for large datasets
2. Sensitive to parameter values
3. Mainly useful only for visualization
4. Distances in low dimensions may not be exactly meaningful
5. Different runs may give different layouts

# Applications of t-SNE

t-SNE is used in:
- image data visualization
- text embedding visualization
- biological data analysis
- cluster exploration
- feature pattern analysis

# When to Use t-SNE?

Use t-SNE when:
- data has many features
- you want to visualize hidden groups
- you want to explore similarity between points

Do not use t-SNE as the main prediction model.

# Difference Between PCA and t-SNE in Use
