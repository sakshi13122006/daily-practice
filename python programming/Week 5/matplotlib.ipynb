```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

x = np.array([12, 23, 32, 37, 44, 49, 51, 56, 59, 64])
y = np.array([10, 14, 19, 25, 30, 35, 40, 43, 51, 59])
z = np.array([70, 65, 61, 58, 50, 48, 42, 37, 28, 15])

plt.title("line plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x, y, color = "red", marker = 'o', mfc = 'blue', ls ='--')
plt.grid()

# markers => <, >, ^, H, o, *
```

```python
plt.title("line plot")
plt.xlabel("x values")
plt.ylabel("z values")
plt.plot(x, z, color = "black", marker = "o", mfc = "red", ls = "--")
plt.grid()
```

```python
plt.title("line plot")
plt.xlabel("x label")
plt.ylabel("y values")
plt.plot(x, y, color ='purple', marker = 'o', mfc = 'blue', ls = '--', label = '2023')
plt.plot(x, z, color = 'black', marker = 'o', mfc = 'red', ls = '--', label = '2024')

plt.grid()
plt.legend()
```

```python
plt.figure(figsize=(16, 9))
plt.subplot(2, 2, 1)
plt.title("line plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x, y, color = 'red', marker = 'o', mfc = 'blue', ls = '--')
plt.grid()

plt.subplot(2, 2, 2)
plt.title("line plot")
plt.xlabel("x values")
plt.ylabel("z values")
plt.plot(x, z, color = 'black', marker = 'o', mfc = 'red', ls = '--')
plt.grid()

```

BAR GRAPH

```python
courses = ['python', 'dart', 'php', 'ruby', 'c++']
students  = [56, 40, 30, 41, 65]

plt.title("Student enrollment")
plt.xlabel('courses')
plt.ylabel("no. of students")
plt.bar(courses, students, color = 'black', width = 0.5, edgecolor = 'red')
plt.grid()
```

PIE CHART

```python
courses = ['python', 'dart', 'php', 'ruby', 'c++']
students = [56, 40, 30, 41, 65]

plt.pie(students, labels = courses, autopct = '%.2f%%', explode = [0.2, 0, 0, 0, 0])
```

```python
# take any four plots from above and put them in a same figure (subplot)

plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
plt.title("line plot")
plt.xlabel("x label")
plt.ylabel("y values")
plt.plot(x, y, color ='purple', marker = 'o', mfc = 'blue', ls = '--', label = '2023')
plt.plot(x, z, color = 'black', marker = 'o', mfc = 'red', ls = '--', label = '2024')

plt.grid()
plt.legend()


plt.subplot(2, 2, 2)
plt.title("Student enrollment")
plt.xlabel('courses')
plt.ylabel("no. of students")
plt.bar(courses, students, color = 'black', width = 0.5, edgecolor = 'red')
plt.grid()



plt.subplot(2, 2, 3)
plt.pie(students, labels = courses, autopct = '%.2f%%', explode = [0.2, 0, 0, 0, 0])


plt.savefig("my_plot.png")
plt.show()
```
