1. What is Seaborn?
  - Seaborn is a Python library for data visualization.
  - Built on top of Matplotlib.
  - Makes charts beautiful, simple, and easy to read.
  - Used widely in data analysis, AI/ML, and engineering projects.


2. Features of Seaborn: 

  - Beautiful default styles – no need to manually style plots.
  - Works with Pandas dataframes – easy for datasets.
  - Supports statistical plots – histograms, regression, boxplots, etc.
  - Can combine multiple plots easily – for comparing data.
  - Automatic color palettes – makes plots visually appealing.


3. Installation
pip install seaborn


4. Importing
import seaborn as sns
Common alias: sns


5. Seaborn Data:

Has built-in datasets for practice, e.g.:

  - tips → restaurant tips

  - iris → flower dataset

  - titanic → passengers survival data



6. Advantages of Seaborn:
    - Easy to use with less code.
    - Prettier plots than Matplotlib.
    - Works directly with dataframes.
    - Good for statistics and ML visualization.


7. Disadvantages:
    - Less flexible than Matplotlib for custom designs.
    - Slower for very large datasets.


8. Basic Syntax:
sns.plot_type(x=…, y=…, data=…, hue=…, style=…, palette=…)

  - x, y → axes

  - data → dataframe

  - hue → color by category

  - palette → color style

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('tips')
data.head()
```

```python
data.shape
```

```python
# HISTOGRAM

sns.histplot(data['total_bill'], kde = False, bins = 15)
plt.title("Total bill distribution")
```

```python
# KDE PLOT

sns.kdeplot(data['total_bill'], fill = True)
plt.title("KDE of total bill ")
```

```python
sns.histplot(data['total_bill'], kde = True, bins = 15)
plt.title('total bill distribution')
```

```python
# boxplot for outlier detection

sns.boxplot(data['tip'])
```

```python
q3 = np.percentile(data['tip'], 75)
q1 = np.percentile(data['tip'], 25)
q3, q1
```

```python
iqr = q3-q1
iqr
```

```python
ub = q3 + iqr * 1.5
lb = q1 - 1.5 * iqr
ub, lb
```

```python
# remove the outliers

x = data[(data['tip'] > lb) & (data['tip'] < ub)]
sns.boxplot(x['tip'])
```

```python
# VIOLIN PLOT -> kde + boxplot

sns.violinplot(data)
```

```python
# stripplot -> jittered scatter plot

sns.stripplot(x = 'day', y = 'total_bill', data = data, jitter = True)
```

```python
# A swarm plot shows every single data point without overlapping.
# strip plot + spacing
# each dot = 1 observation

sns.swarmplot(x = 'day', y = 'total_bill', data = data)
```

```python
# scatter

sns.scatterplot(x = 'total_bill', y ='tip', data = data, hue = 'day')
```

```python
# regplot = Regression plot

sns.regplot(x  = 'total_bill', y = 'tip', data = data)
```

```python
# heatmap very imp

corr = data.corr(numeric_only =True)
corr
```

```python
'''ArithmeticErrorA heatmap is a color-coded table that shows values clearly using colors.

Here:

corr → a correlation matrix (or any 2D data)

annot=True → writes the numbers on the cells

cmap='coolwarm' → colors the cells from cool (blue) to warm (red)'''

sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```

```python
'''This function plots the counts of each unique value in a column (basically a bar chart showing frequency).

Example: if a column has ['Lunch', 'Dinner', 'Dinner', 'Lunch', 'Dinner'], it will plot a bar for Lunch = 2, Dinner = 3.'''

sns.countplot(data['time'], color = 'purple')
```

```python
"""This function plots pairwise relationships in a dataset.

It creates a grid of scatter plots for every numerical column against every other numerical column."""

sns.pairplot(data, hue = 'smoker')
```

```python
"""jointplot()

This function plots a relationship between two variables (x vs y).

It shows:

Scatter plot (or other kind) of x vs y in the main area.

Histograms for x (on top) and y (on the right) to show their distributions."""

sns.jointplot(x = 'tip', y = 'total_bill', data = data, hue = 'day')
```
