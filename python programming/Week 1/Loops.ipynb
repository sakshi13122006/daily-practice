********* LOOPS  ********
- Loops are used to repeat block of codemultiple times until the condition is satisfied.

There are two types of Loops used in Python:
          1. while loop
          2. for loop 
          3. nested loop 

    1. while loop
       - while loop executes a set of statement as long as a condition is true.
         (This loop is used when number of iterations is not known and depends on a condition.)

```python
# While loop

i = 1
while i < 6:
    print(i)
    i += 1
```

2. for loop 
        - we use for loop repeat a block of code a fixed number of times or to iterate over sequence like lists, tuple, string, range
        (when we know the number of iterations.)

```python
for i in range(3):   # for loop
    print("Hello")
```

```python
name = "Sakshi"
for ch in name:
     print(ch)    
```

3. nested loop 
         - nested loop means loop inside loop

```python
for row in range(1, 4):
   for star in range(1, 4):
        print("*", end=" ")
        print()
```

- BREAK Statement
          - Using break statement we can stop the execution of loop even if the while condition is true.

```python
i = 1
while i < 6:
    print(i)
    if (i == 3):
        break
    i += 1
```

-CONTINUE Statement
           - Using continue statement we can stop current iteration and run next one.

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        i += 1
        continue
    i += 1
```

-The else statement:
            - using else statement we can run block of code once when the condition no longer is true.

```python
i = 1
while i < 6:
     print(i)
     i += 1
else:
    print("i is no longer less than 6")
```
