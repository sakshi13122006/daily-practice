4. TYPE CASTING
   - converting one data type into another.

   TYPES OF TYPE CASTING
   4.1 Implicit CASTING
      - Automatically done by Python.
      - Smaller data type conveerted into larger data type.

      - ex;

           x = 5   #int
           y = 2.5 # float
           z = x + y #x is implicitly convertedto float
           priny(z)

   4.2 Explicit CASTING
     - Done by programmer using functions like int(), float(), str()

     -ex;
         x = 5.7
         y = int(x)   #float to int
         z = str(y)   #int to string
         print(y , z)



```python
# Implicit Type casting

x = 5   #int
y = 2.5 # float
z = x + y #x is implicitly converted to float
print(z)
print(type(z))
```

```python
# Explicit Typecasting
x = 5.7
y = int(x)   #float to int
z = str(y)   #int to string
print(y , z)
print(type(y))
print(type(z))
```

5. OPERATORS
    - An operator is a symbol used to perform operations on variables that produces result.

    5.1 arithemetic operator

       OPERATORS      Meaning          Example
          +           Addition          5+3=8
          -           Substraction      5-3=2
          *           Multiplication    5*3=15
          /           Division          5/2=2.5
          //          Floor Division    5//2=2
          %           Modulus           5%2=1

```python
# Example of Arithmetic Opeartor

a = 100
b = 70

print("Addition: ", a + b)
print("Sustraction: ", a - b)
print("Multiplication: ", a * b)
print("Division: ", a / b)
print("Floor Division: ", a // b)
print("Modulus: ", a % b)
print("Exponent: ", a ** b)
```

5.3 Logical OPERATORS

         OPRATORS         MEANING                    
          and             true if both are true     
          or              true if any is true
          not             reverse boolean value     


```python
# Example of Logical Operator

a = 5
b = 3
print(a > 0 and b > 0)
print(a > 0 or b < 0 )
print(not(a > b))
```

5.5 BITWISE OPERATORS 
         OPERATOR     MEANING          EXAMPLE
          &           AND              5&3=1
          ^           XOR              5^3 = 6
          ~           NOT              ~5=-6
          <<          left shift       5<<1=10

```python
# Example of Bitwise Operator

a = 5
b = 3
print(a & b)
print(a | b)
print(a ^ b)
print(~a)
print(a << 1)
print(a >> 1)
```

9. SCOPE OF VARIABLES
      - Scope of a variable referes to the region of a program whre that variable is recognized and accessible. Python has two main types of variable scope.

      LOCAL SCOPE:
            - Variables defined inside a function are only accessible within that function.

      GLOBAL SCOPE:
            - Variables defined outside any function are accessible throughput the program.


```python
# Scope of variable

x = 10  #global variable
def show():
     x = 5  #local variable
     print("Inside function, x:", x)  

show()
print("Output function, x:", x) 

```
