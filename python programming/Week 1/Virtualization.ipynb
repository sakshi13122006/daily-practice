*** Virtualization concept ***
    
1. Virtualization :
    Virtualization means creating a separate environment inside our computer so that software runs without affecting other software.

2. Why we use virtualization?
    - To avoid software version problems
    - To keep projects separate
    - To run same project on different systems easily

3.Types (Basic)
    Docker – runs applications in containers
    Virtual Environment (venv / conda) – used mainly in Python

4. How I craeted my Virtual Environment :
    - Firstly I opened my project folder in VS code.
    - I pressed Ctrl + Shift + P.
    - I Searched for Python: Create Environment.
    - then Selected venv.
    - I selected Interpreter.
    (After this, VS Code Automatically created a folder named .venv(virtual environment) and also automatically activated the environment).

5. Installation of libraries inside virtual environment:
    - I created a file named requirements.txt in my project folder.
    - Inside that file, I wrote library names one by one.
    - Example:
              numpy
              pandas
              matplotlib
              flask
              django
    - Opened terminal and ran: pip install -r requirements.txt
    (This command installed all libraries at once inside my virtual environment.)

6. Git & GitHub (Connecting project to GitHub)
   After creating virtual environment and project file, I connected project to GitHub.

   Creation of repository on GitHub
    - Created GitHub account.
    - logged in to my GitHub account.
    - Clicked on New Repository.
    - Entered Repository name and selected Public repository.
    - Did not add README file.
    - Then I clicked on Create repository.
    - GitHub generated a repository URL.
    - I copied that URL.

    Git setup(command prompt)
    - git config --global user.name "Sakshi" 
    - git config --global user.email "Myemail@gmail.com"

    Upload repository to GitHub
    - Initialize Git:
                      git Init
    - Check Status:
                      git Status
    - Add files:
                      git add .
    - commit:
                      git commit -m "message"
    - connect GitHub Repo:
                      git remote add origin <repo_copied url>
    -Push code:
                      git branch -M main
                      git push -u origin main

    Daily Push Commands:
                         git add .
                         git commit -m
                         git push
