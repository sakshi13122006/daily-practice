************ DECISION MAKING IN PYTHON ************
   
      - Decision making allows the program to execute different blocks of code based on conditions.

      TYPES:
          1. if statement 
          2. if-else Statement 
          3. if-elif-else statement
          4. Nested if

        1. if statement:
             - Executes a block of code only if the condition is true.

             Syntax:
                   if condition:
                        # block of code to execute if condition is true


```python
# Example of if Statement

age = 18
if age >= 18:
    print("you are eligible to vote.")
```

2. if-else Statement:
                - executes one block if the condition is true and another block if the condtion is false.

               Syntax:
                      if condition:
                              # block of code if condition is true
                      else:
                              # block of cde is the condition is false


```python
# Exapmle of if-else Statement

age = 16
if age >= 18:
    print("you can vote")
else:
    print("you cannot vote")
```

3. if-elif-else statement:
                  - Used when thre are multiple conditions to check. Python executes the first true contion if finds.

               syntax:
                     if condition1:
                             # block of code if condition1 is true

                     elif condition2:
                              # block of code if condition2 is true
             
                     else:
                              #block of code if the all conditions are false

```python
# Example of if-elif-else statement

marks = 75
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print ("Grade C")
else:
    print("Fail")
```

 4. Nested if Statement:
                  - An if or if...else statement inside another if statement.

                  Syntax:
                        if condition1:
                            if condition2:
                                 # block of code if both conditions are true


```python
num = 10
if num > 0:
    if num % 2 == 0:
     print("Positive even number")
                         
```
