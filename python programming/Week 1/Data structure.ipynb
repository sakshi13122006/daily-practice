   *************** DATA STRUCTURE IN PYTHON ***************
  
     - Data structure are ways of organizing and storing data so you can access and modify it efficiently.
     - It provides built in data structures like list, tuple, set and dictionaries.

     1. LIST 
       - A list stores a sequence of items in order. You can change elements after creation.
       - Features:
                  - Ordered
                  - mutable
                  - allows duplicates

       - Operations:
                  - sort 
                  - reverse
                  - remove
                  - update
                  - append
                  - insert
                  - extend 

```python
nums = [10, 20, 30, 40]
nums.append(50)
print(nums[0])  
         
```

```python
list = ["apple","banana","orange","cherry"]
list.insert(1,"kiwi")
list.sort()
list.remove("banana")
list[1] = "strawberry"
list.extend(["banana","papaya","guava"])
print(list)
```

  2. Tuple:
        - tuple is like list but it is immutable we cannot modified it after it's creation.
        - Features:
                  - ordered 
                  - immutable 
                  - allows duplicates

         - Operations:
                      - len (we can fint out the length)
                      - count (we can get how many times an item appears)
                      - index (we can also get an item by an index)

```python
point = (3, 4)
print(point[1]) 
```

```python
t1 = (10,20)
t2 = (30,40,50)
t3 = t1 + t2
print(t3)
```

```python
t = (10,20,30,40,50)
if 20  in t:
    print("element found")
```

3. set :
         -  A set is a collection of unordered unique items.
         - Feature:
                   - unordered
                   - mutable
                   - no duplicates

         - Opeartions:
                   - add
                   - remove
                   - discard
                   - union
                   - intersection
                   - difference

```python
fruits = {"apple", "banana", "orange"}
fruits.add("mango")
print(fruits)
```

 4. dictionary :
          - a dictionary is collection of key-value pair. but in this key must be unique, value can be repeat.
          - Features:
                   - unordered
                   - mutable
                   - very fast look up by key

         - Operations:
                  - add
                  - update
                  - remove
                  - we can also get keys ,avlues and pair by using their words.

```python
dict = {"name": "Asha", "roll-no": 22 }
print(dict["name"])              
```
